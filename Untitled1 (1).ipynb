```python
# dfd=dfd[dfd['condition'].isin(['Diabetes, Type 1','Diabetes, Type 2'])]
# dfdt=dfd[dfd['rating']>6]
# # dfdt=dfd.set_index('condition')
# # dfdt.head(50)
# # dft.loc['Diabetes, Type 1']
import matplotlib.pyplot as plt

# # dfdt['rating'].head(40).plot(kind='bar')
# # dfdt.info()
# dfdt
import pandas as pd
!pip install pandas-profiling
from pandas_profiling import ProfileReport

df=pd.read_csv(r"C:\Users\Shashwat\Downloads\archive (1)\drugsComTest_raw.csv")
dfd=df[df['condition'].isin(['Diabetes, Type 1','Diabetes, Type 2'])]
dft=pd.read_csv(r"C:\Users\Shashwat\Downloads\archive (1)\drugsComTrain_raw.csv")
dft=dft[dft['condition'].isin(['Diabetes, Type 1','Diabetes, Type 2'])]
dftr=dfd.merge(dft,how='outer')
final=dftr.drop(columns=['date','usefulCount','uniqueID'])
final.set_index('condition')
final.info()
final.head(50)
type1=final[final['condition'].isin(['Diabetes, Type 1'])]
type2=final[final['condition'].isin(['Diabetes, Type 2'])]
pd.concat([type1,type2])
type1.info()
type2.info()

type2report=ProfileReport(type2,explorative=True)
type2report.to_file('type2report.html')
```

```python
type22=type2.drop(columns=['condition']).set_index('drugName')
type22.drop_duplicates()
type22=type22.groupby('drugName').agg({'rating':['mean','max']})

type22['rating'].head(30).plot(kind='bar')
type22.info()
```
